Analysing data from data frame *********************

```python
#number of reported cases and deaths
total_cases=covid_df.new_cases.sum()
total_deaths=covid_df.new_deaths.sum()
```

```python
#overall deaths rate
death_rate=total_deaths/total_cases
```

************Grouping and aggregation *********************************************

```python
covid_year_df=covid_df.groupby('year')[['new_cases', 'new_deaths', 'new_tests']].sum()
```

```python
covid_year_means_df=covid_df.groupby('year')[['new_cases', 'new_deaths', 'new_tests']].mean()
```

```python
# Merging data from multiple sources
locations_df=pd.read_csv('locations.csv')
```

```python
locations_df
```

```python
locations_df[locations_df.location=="Italy"]
```

```python
covid_df['location']="Italy"
```

```python
covid_df
```

```python
merged_df=covid_df.merge(locations_df, on="location")
```

```python
merged_df
```

```python
merged_df['deaths_per_million']=merged_df.new_deaths*1e6/merged_df.population
```

```python
merged_df
```
